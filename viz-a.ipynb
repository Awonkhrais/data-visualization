```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Set seaborn

```python
sns.set()
```

```python
sns.get_dataset_names()
```

## Getting The DataFrames

```python
df = sns.load_dataset('penguins')
df.head()
```

## Visualizing Data (Bar Plot)


```python
sns.barplot(x='bill_length_mm', y='body_mass_g', data=df.head(10))
```

## Visualizing Data (Relational Plot)

```python
sns.relplot(x='bill_length_mm', y='body_mass_g', data=df)
```

## Visualizing Data (Linear Regrusion)

```python
sns.lmplot(x='bill_length_mm', y='body_mass_g', data=df)
```

# Conclusions(summary): 
The realtion here shows that the body mass increase when the bill length increase
